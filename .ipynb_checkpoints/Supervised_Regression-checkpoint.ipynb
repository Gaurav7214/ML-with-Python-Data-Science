```python
# Total count(sales) as a function of temp
# Independant variable = Temp
# Dependant variable = Count

# y = mx + c
# cnt = m(temp) + c
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()  # LinearRegression() object
```

```python
#Fit and #Predict
```

# Simple Linear Regression

```python
df["temp"].shape
```

```python
df["temp"].values
```

```python
df["temp"].values.reshape(-1,1)
```

```python
df["temp"].values.reshape(-1,1).shape
```

```python
lr.fit(df["temp"].values.reshape(-1,1),df["cnt"].values.reshape(-1,1))
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.predict([[0.45]])
```

```python
lr.predict([[0.15]])
```

```python
lr.predict([[0.344167]])
```

```python
plt.scatter(df["temp"],df["cnt"])
```

```python
m = lr.coef_
c = lr.intercept_
```

```python
plt.scatter(df["temp"],df["cnt"])
x = df["temp"]
```

```python
x
```

```python
import numpy as np
```

```python
X = np.array(x)
```

```python
X = X.reshape(-1,1)
X
```

```python
Y = ((m*X) + c).reshape(-1,1)
Y
```

```python
plt.plot(X, Y)
```

```python
plt.scatter(df["temp"], df["cnt"])
plt.plot(X, Y, color="red")
```

# Multipal Linear Regression

```python
df.columns
```

```python
df[["temp","hum","casual"]]
```

```python
lr.fit(df[["temp","hum","casual"]], df["cnt"].values.reshape(-1,1))
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.coef_[0,0]
```

```python
c = lr.intercept_
m1 = lr.coef_[0,0]
m2 = lr.coef_[0,1]
m3 = lr.coef_[0,2]
X1 = np.array(df["temp"]).reshape(-1,1)
X2 = np.array(df["hum"]).reshape(-1,1)
X3 = np.array(df["casual"]).reshape(-1,1)
Y = (m1*X1) + (m2*X2) + (m3*X3)
```

```python
plt.scatter(df["temp"], df["cnt"])
```

```python
plt.scatter(df["hum"], df["cnt"])
```

```python
plt.scatter(df["casual"], df["cnt"])
```

```python
lr.predict([])
```
