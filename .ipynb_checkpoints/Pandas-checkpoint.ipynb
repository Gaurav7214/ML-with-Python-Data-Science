# (III) Skewness

```python
# Measure of symmetry of a distribution compared to a standard normal distribution.
# +ve: right skewed(mean is to the right of mode/median). Long tail in the +ve direction.
# 0: Symmetric.
# -ve: left skewed(mean is to the left of mode/median). Long tail in the -ve direction.
df.skew()
```

# (IV) Kurtosis

```python
# Measure of flatness or peakedness of a distribution compared to a normal distribution.
# +ve: Leptokurtosis (sharper/spikier peak compared to the normal distribution.)
# 0: Mesokurtic (normal distribution.)
# -ve: Platykurtic (flatter peak compared to the normal distribution.) eg. Uniform distribution.
df.kurtosis()
```

# Outliers

```python
dict_ = {"Names": ['John', 'Paul', 'Mark', 'Sarah', 'Morgan', 'Mike', 'Ram', 'Jennifer', 'Amy', 'Chris', 'Mitchel', 'Alex'], "Age": [25, 45, 12, 95, 6, 21, 34, 24, 41, 31, 19, 16]}
dff = pd.DataFrame(dict_)
dff
```

```python
# Sarah's age is changing the mean, standard deviation significantly and makes error
# Compute z-score
```

```python
from scipy import stats
```

```python
stats.zscore(dff["Age"]) # z-score: deviation from mean
```

```python
# For two sided test the z-scores for the thresholds are: 95 percentile: 1.96, 99 percentile: 2.58
# z-score of sarah's age is greater than 99 percentile
```

```python
dff = dff[abs(stats.zscore(dff["Age"])) < 1.96]
dff = dff.reset_index(drop=True) # Reset index
dff
```

```python
# What if you want to apply this to all columns?
#dff = dff[(abs(stats.zscore(x))<1.96).all(axis=1)]
```
